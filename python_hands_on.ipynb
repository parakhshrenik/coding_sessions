# Understanding loops

**What are loops?** ->  _loops are used to repeat a block of code_

Types of loops in Python: 
- for loop - to repeat a block of code for a definite number of times
- while loop - to repeat a block of code till a condition is met

### Examples of `for` loop
`for` loops can be used to iterate over sequences such as lists, tuples, string, etc.

Example: Suppose we've the following languages
```
languages = ['Swift', 'Python', 'Go', 'JavaScript']
```
And we want to print each of the language, we don't want to do something like
```
print (languages[0])
print (languages[1])
print (languages[2])
...
```
So, instead we'll use loops like below:

````python
# Suppose we've the following languages
languages = ['Swift', 'Python', 'Go', 'JavaScript']

for language in languages:
    print (language)
````

### Examples of `while` loop
`while` loop is used to run a block code until a certain condition is met.  
For example keep adding the numbers a user enters till a `0` is received

````python
total = 0
# Take a user input
input_num = int(input('Enter a number: '))
while input_num != 0:
    total += input_num    # total = total + number
    input_num = int(input('Enter a number to add, please enter 0 to stop... '))
    
print('total =', total)
````

# break and continue

- `break` - a break is used to stop iteration over any loop
- `continue` - is used to skip that item and move to next iteration

````python
fruits = ["apple", "banana", "orange", "mango"]
my_fav_fruits = "mango"

for fruit in fruits:
    if fruit == my_fav_fruits:
        print (f"my_favourite fruit is {fruit}")  # Note the use of f-strings
        break
````

````python
fruits = ["apple", "banana", "orange", "pomegranate", "mango"]
disliked_fruit = "pomegranate"

for fruit in fruits:
    if fruit == disliked_fruit:  # == is a relational operator to check if both values are equal or mot
        continue
    print (f"I don't mind {fruit}...")  # Note the use of f-strings
    print (f"Place an order for {fruit}\n______________________")
        
````

# Funtions 
A function is a block of code which runs only when it is called  
- You can pass arguments to a function
- a function is defined using the `def` keyword (_usually referred to as function definition_)
- a function isn't executed when it is defined
- a function is only executed when it is called (_usually referred to as function call_)

## Arguments/Parameters:
Arguments are the information that is passed to a function which maybe required to perform and run the block of code written within the function
```From a function's perspective:
A parameter is the variable listed inside the parentheses in the function definition.
An argument is the value that is sent to the function when it is called.
```

````python
def welcome(student_name):
    print (f"Welcome to the session {student_name}")
````

````python
welcome("Siddharth")
````

````python
def welcome(students: list, session_count: str):
    for student in students:
        print (f"Welcome to the session number {session_count} :  {student}")
````

````python
welcome(
    students = ["foo", "bar"],
    session_count = 2
)
````

### Default parameter value
A default parameter value is where if that argument is not sent, a default parameter is used

````python
def welcome_with_country(name, country = "India"):
    print (f"Welcome {name} from {country}")
    
welcome_with_country("Rohit Sharma")
````

## Let's understand it with an example: 
We're going to use the same example as used in the prev session  
We'll create a class Car. It'll have certain attributes and functions, that you'd expect a car to have

````python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def drive(self):
        print("The car is driving")

    def stop(self):
        print("The car is stopping")
````

This class defines three properties: `make`, `model`, and `year`. It also defines two methods: `drive()` and `stop()`

Let's now create an object of the class

````python
# In other words, the Car class is like a template for creating Car objects. 
#The my_car object is an instance of the Car class. It has the properties and behaviors defined by the Car class.
my_car = Car("Toyota", "Corolla", 2023)
````

````python
# Let's take another example, before going back to the example of car
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def circumference(self):
        return 2 * (self.length + self.width)

# This makes sense, right? 
# Every circle has an attribute radius, and we've implemented methods to calculate area and circumference of a circle
# if we create an object of a class circle
my_rectangle = Rectangle(5, 10)
````


> How many dimensions are there in a rectangle

That's something we can all answer, without needing an object, all rectangles have 2 dimensions, and the objects need not have an individual copy of this attribute, they can all share it from the class itself. Hence we can define `dimensions` as a `class variable`

````python

class Rectangle:
    dimensions = 2
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def circumference(self):
        return 2 * (self.length + self.width)
    
my_rectangle = Rectangle(5, 10)
print (type(my_rectangle))  # if anyone is curious __main__ is the name of the environment where top-level code is run
````

````python
print (f"printng from class variable, number of dimensions in a rectangle is: {Rectangle.dimensions}")
print (f"printng from instance variable, number of dimensions in my rectangle is: {my_rectangle.dimensions} \n\n")

print (f"Using the method, Area of my rectangle is : {my_rectangle.area()}")
````

# Inheritance

Inheritance is when you have a main class (parent) that has certain characteristics and behaviors, and you want to create another class (child) that is similar but with some extra or modified features. The child class can "inherit" the properties and methods of the parent class, so you don't have to rewrite everything.

We'll try to understand this going back to our example of `class Car`  
Redefine the class car

````python
class Car:
    no_of_wheels = 4
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def drive(self):
        print(f"Your car {self.make} {self.model} is driving")

    def stop(self):
        print("The car is stopping")
````

````python
class ElectricCar(Car):
    def __init__(self,make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity
        
    def charging(self):
        print("Your electric car is charging")
````

````python
my_old_car = Car("Maruti", "Alto", 2016)
my_fancy_new_car = ElectricCar("Tesla", "Roadster", 2022, "200 kWh")
````

````python
my_old_car.drive()
my_fancy_new_car.drive()
````

````python
my_old_car.no_of_wheels
````

````python
my_fancy_new_car.no_of_wheels
````

````python
my_fancy_new_car.charging()
````

````python
my_old_car.charging()
````

````python
class ElectricCar(Car):
    def __init__(self,make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity
        
    def charging(self):
        print("Your electric car is charging")
    
    # Overrides the behaviour of parent class
    def drive(self):
        print(f"Your Electic Car {self.make} {self.model} is driving. Enjoy the ride!!")
# Re-create an object of this child class
my_fancy_new_car = ElectricCar("Tesla", "Roadster", 2022, "200 kWh")
````

# Abstraction

Abstract classes in OOPs provide a blueprint with method declarations that must be implemented by their concrete subclasses.
As the name suggests, its meant to achieve abstraction. 
A very important thing to keep in mind
>An abstract class is a class that cannot be instantiated. It can only be inherited from


````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        raise NotImplementedError()

    @abstractmethod
    def perimeter(self):
        raise NotImplementedError()
````

````python
x = Shape()
````

````python
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side
````

````python
x = Square(4)
````

````python
class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side
    def perimeter(self):
        return 4 * self.side
````

````python
my_square =  Square(5)
````

````python
print(f"Side of the square is {my_square.side}")
print (f"Area of the square is {my_square.area()}")
print (f"perimeter of the square is {my_square.perimeter()}")
````
